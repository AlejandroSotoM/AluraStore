### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Ingreso_total_tienda1 = sum(tienda['Precio'])
print("Ingreso total de la tienda 1: ", Ingreso_total_tienda1)
Ingreso_total_tienda2 = sum(tienda2['Precio'])
print("Ingreso total de la tienda 2: ", Ingreso_total_tienda2)
Ingreso_total_tienda3 = sum(tienda3['Precio'])
print("Ingreso total de la tienda 3: ", Ingreso_total_tienda3)
Ingreso_total_tienda4 = sum(tienda4['Precio'])
print("Ingreso total de la tienda 4: ", Ingreso_total_tienda4)
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
print("Ventas por categoría en Tienda 1:")
display(ventas_por_categoria_tienda1)

ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
print("\nVentas por categoría en Tienda 2:")
display(ventas_por_categoria_tienda2)

ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
print("\nVentas por categoría en Tienda 3:")
display(ventas_por_categoria_tienda3)

ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
print("\nVentas por categoría en Tienda 4:")
display(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print("Calificación promedio de la Tienda 1:", calificacion_promedio_tienda1)

calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print("Calificación promedio de la Tienda 2:", calificacion_promedio_tienda2)

calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print("Calificación promedio de la Tienda 3:", calificacion_promedio_tienda3)

calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print("Calificación promedio de la Tienda 4:", calificacion_promedio_tienda4)
```

# 4. Productos más y menos vendidos

```python
# Tienda 1
productos_vendidos_tienda1 = tienda['Producto'].value_counts()
productos_mas_vendidos_tienda1 = productos_vendidos_tienda1.head(10)
productos_menos_vendidos_tienda1 = productos_vendidos_tienda1.tail(10)

print("Productos más vendidos en Tienda 1:")
display(productos_mas_vendidos_tienda1)
print("\nProductos menos vendidos en Tienda 1:")
display(productos_menos_vendidos_tienda1)

# Tienda 2
productos_vendidos_tienda2 = tienda2['Producto'].value_counts()
productos_mas_vendidos_tienda2 = productos_vendidos_tienda2.head(10)
productos_menos_vendidos_tienda2 = productos_vendidos_tienda2.tail(10)

print("\nProductos más vendidos en Tienda 2:")
display(productos_mas_vendidos_tienda2)
print("\nProductos menos vendidos en Tienda 2:")
display(productos_menos_vendidos_tienda2)

# Tienda 3
productos_vendidos_tienda3 = tienda3['Producto'].value_counts()
productos_mas_vendidos_tienda3 = productos_vendidos_tienda3.head(10)
productos_menos_vendidos_tienda3 = productos_vendidos_tienda3.tail(10)

print("\nProductos más vendidos en Tienda 3:")
display(productos_mas_vendidos_tienda3)
print("\nProductos menos vendidos en Tienda 3:")
display(productos_menos_vendidos_tienda3)

# Tienda 4
productos_vendidos_tienda4 = tienda4['Producto'].value_counts()
productos_mas_vendidos_tienda4 = productos_vendidos_tienda4.head(10)
productos_menos_vendidos_tienda4 = productos_vendidos_tienda4.tail(10)

print("\nProductos más vendidos en Tienda 4:")
display(productos_mas_vendidos_tienda4)
print("\nProductos menos vendidos en Tienda 4:")
display(productos_menos_vendidos_tienda4)
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
print("Costo de envío promedio de la Tienda 1:", costo_envio_promedio_tienda1)

costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
print("Costo de envío promedio de la Tienda 2:", costo_envio_promedio_tienda2)

costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
print("Costo de envío promedio de la Tienda 3:", costo_envio_promedio_tienda3)

costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
print("Costo de envío promedio de la Tienda 4:", costo_envio_promedio_tienda4)
```

## Visualizar ingresos totales por tienda

### Subtask:
Crea un gráfico de barras para comparar los ingresos totales de cada tienda.


**Reasoning**:
Create a DataFrame with the total revenue for each store and then generate a bar plot to visualize the revenue comparison between stores.



```python
import matplotlib.pyplot as plt

ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [Ingreso_total_tienda1, Ingreso_total_tienda2, Ingreso_total_tienda3, Ingreso_total_tienda4]
})

plt.figure(figsize=(10, 6))
plt.bar(ingresos_totales['Tienda'], ingresos_totales['Ingreso Total'], color=['blue', 'green', 'red', 'purple'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.show()
```

## Visualizar ventas por categoría

### Subtask:
Crea un gráfico de barras apiladas o agrupadas para mostrar la distribución de ventas por categoría en cada tienda.


**Reasoning**:
Concatenate the dataframes, group by store and category, and then pivot the data to prepare for plotting.



```python
import matplotlib.pyplot as plt

tienda['Store'] = 'Tienda 1'
tienda2['Store'] = 'Tienda 2'
tienda3['Store'] = 'Tienda 3'
tienda4['Store'] = 'Tienda 4'

all_stores = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

category_sales = all_stores.groupby(['Store', 'Categoría del Producto']).size().unstack(fill_value=0)

display(category_sales)
```

**Reasoning**:
Generate a stacked bar plot from the pivoted data to visualize the distribution of sales by category for each store, including title, labels, and legend.



```python
category_sales.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribución de Ventas por Categoría y Tienda')
plt.xlabel('Tienda')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=0)
plt.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Visualizar calificación promedio por tienda

### Subtask:
Crea un gráfico de barras para comparar la calificación promedio de cada tienda.


**Reasoning**:
Create a DataFrame with the average ratings and generate a bar chart to visualize the average rating for each store.



```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

average_ratings = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
})

plt.figure(figsize=(10, 6))
plt.bar(average_ratings['Tienda'], average_ratings['Calificación Promedio'], color=['blue', 'green', 'red', 'purple'])
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.show()
```

## Visualize costs

### Subtask:
Visualize shipping costs per store.


**Reasoning**:

```python
import matplotlib.pyplot as plt

average_shipping_costs = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2, costo_envio_promedio_tienda3, costo_envio_promedio_tienda4]
})

plt.figure(figsize=(10, 6))
plt.bar(average_shipping_costs['Tienda'], average_shipping_costs['Costo de Envío Promedio'], color=['blue', 'green', 'red', 'purple'])
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.show()
```
